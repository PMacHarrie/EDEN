```python
def r_to_j(r):
    myJ_s = ''
    line_num=1
    for n in r:
        #print(line_num, n)
        myJ_s += n.decode("'iso-8859-1'")
        
#    print(myJ_s)
    myJ = json.loads(myJ_s)
#    print(json.dumps(myJ_s, indent=1))
    return myJ
```

```python
def do_r(req, page_size=2000, page_num=0,debug=False, attrs=[]):
    myJ = {}
    rJ = {'count': 0, 'g' : {}}
    keepGoing=1

    req += '&page_size='+str(page_size)+'&page_num='

    while keepGoing:
    
        page_num += 1
        if page_num > 499:
            keepGoing = 0
        
        r = requests.get(req + str(page_num) )
        if debug:
            print(r, page_num, datetime.datetime.now())

        myJ = r_to_j(r)
#        print('myj=', json.dumps(myJ, indent=1))
        if len(myJ['items']) == 0:
            keepGoing=0
            continue

        rJ['count']+= len(myJ['items'])
        outJ={}
        for j in myJ['items']:
            printIt=0
            if printIt:
                print(json.dumps(j, indent=1))
                printIt=0
            
            outJ={}
            for a in attrs:
                #print(a,'a=')
                if a in j['umm']:
                    outJ[a]=j['umm'][a]
                
            rJ['g'][j['meta']['native-id']]=outJ

    return rJ
```
