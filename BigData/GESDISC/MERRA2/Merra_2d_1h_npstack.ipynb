```python
def do_it(m_str, m):
    #print (d)
    outs=[]
    vars={}

    for d in m:
        #print (d)
        ds=m[d]['ds']
        for v in ds.variables:
            if v in vars:
                print ('duplicate variable: ', v, ' skipping ....')
            else:
                ds.set_auto_mask(False)
                if '_FillValue' in ds[v].ncattrs():
                    tmp=ds[v][:]
                    tmp[tmp == ds[v]._FillValue] = np.nan
                    vars[v]=tmp[:]
                else:
                    vars[v]=ds[v][:]
        print (d)
        
# Normalize lat/lon
    vm_lon, vm_lat = np.meshgrid(vars['lon'], vars['lat'])
    #print (vm_lat.shape, vm_lon)

# Normalize time

    v_time = nc.num2date(vars['time'][:],ds['time'].units, only_use_python_datetimes=True)
    #print (v_time)
    vm_time=np.zeros(361*576, dtype='datetime64[s]')
#    vm_time[:]= v_time

    #print (vm_time[0])

    t_size=vars['time'].size
    for t in range(t_size):
        #print (t)
        stack_this = []
        ctr=0
        for v in sorted(vars):
            #print (ctr, v, vars[v].shape)
            ctr+=1
            if v == 'time':
                vm_time[:] = v_time[t]
                print ('vmtime', vm_time.shape)
                stack_this.append(vm_time.astype('float'))
            elif v == 'lon':
                stack_this.append(vm_lon.flatten()[:])
            elif v == 'lat':
                stack_this.append(vm_lat.flatten()[:])
            else:
                stack_this.append(vars[v][t].flatten()[:])
        #print (stack_this)
        x=np.dstack( (stack_this) )
        fname = '/home/ec2-user/merra2-2d_hourly_' + m_str + '_' + str(t) + '.tsv.gz'
        #print ('fname=', fname)
        np.savetxt(fname, x[0], fmt='%.12g', delimiter='~', newline="\n")
        outs.append(fname)
        
    return outs
```
